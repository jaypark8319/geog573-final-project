```python
sido_mapping = {
    '서울특별시': '서울', '경기도': '경기', '강원특별자치도': '강원','부산광역시': '부산', '대구광역시': '대구', '인천광역시': '인천', '광주광역시': '광주'
    , '세종특별자치시': '세종',  '대전광역시': '대전', '충청북도': '충북', '충청남도': '충남', '울산광역시': '울산',
    '전북특별자치도': '전북', '전라남도': '전남', '경상북도': '경북', '경상남도': '경남', '제주특별자치도': '제주'
}

result['SIDO'] = result['SIDO'].map(sido_mapping).fillna(result['SIDO'])

current_sido = None
for idx in result.index:
    if pd.notna(result.loc[idx, 'SIDO']):
        current_sido = result.loc[idx, 'SIDO']
    else:
        result.loc[idx, 'SIDO'] = current_sido
```

```python
protected_gu = ['동구', '서구', '남구', '북구', '중구']
suffixes = '갑을병정무'

def is_protected_name(name):
    if name in protected_gu:
        return True
    if any(name == gu + s for gu in protected_gu for s in suffixes):
        return True
    gu_combo = '|'.join(protected_gu)
    pattern = fr'^({gu_combo})({gu_combo})[{suffixes}]?$'
    return bool(re.match(pattern, name))

def remove_gu_seoul(name):
    if is_protected_name(name):
        return name
    name = re.sub(r'구(?=[갑을병정무]$)', '', name)
    name = re.sub(r'구$', '', name)
    return name

def remove_gu_other(name):
    if is_protected_name(name):
        return name
    name = re.sub(r'구(?=[갑을병정무]$)', '', name)
    name = re.sub(r'구$', '', name)
    return name

def remove_gun_si(name):
    if pd.isna(name):
        return name
    first_char = name[0]
    rest = name[1:].replace('시', '').replace('군', '')
    return first_char + rest

def clean_polling_place(name, sido):
    if pd.isna(name):
        return name

    # Step 0: Handle hardcoded special cases
    if name == '동구군위군갑':
        return '동구군위갑'
    if name == '동구군위군을':
        return '동구군위을'

    # Step 1: Handle '구'
    name = remove_gu_seoul(name) if sido == '서울' else remove_gu_other(name)

    # Step 2: Remove '시' and '군' except at start
    name = remove_gun_si(name)

    # Step 3: Final special case
    if name == '포항남구울릉':
        name = '포항남울릉'
    if name == '포항북':
        name = '포항북구'
    if name == '세종특별자치갑':
        name = '세종갑'
    if name == '세종특별자치을':
        name = '세종을'

    return name

# Apply to DataFrame
result['SGG'] = result.apply(lambda row: clean_polling_place(row['SGG'], row['SIDO']), axis=1)
```
